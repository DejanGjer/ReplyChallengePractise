```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style='darkgrid', color_codes=True)
print('Setup complete')
```

```python
mountain = 2000
def polje(c):
    if c == '#':
        return mountain
    elif c == '~':
        return 800
    elif c == '*':
        return 200
    elif c == '+':
        return 150
    elif c == 'X':
        return 120
    elif c == '_':
        return 100
    elif c == 'H':
        return 70
    else:
        return 50
```

```python
filename = '5_oceania.txt'
lines = open(filename,'r').readlines()
N,M,C,R = [int(val) for val in lines[0].split()]
l = 1
xc = []
yc = []
rc = []
for i in range(C):
    x,y,r = [int(val) for val in lines[l].strip().split()]
    xc.append(x)
    yc.append(y)
    rc.append(r)
    l += 1
xc = np.array(xc)
yc = np.array(yc)
rc = np.array(rc)
mapa = []
for i in range(M):
    mapa.append(list(map(polje, lines[l].strip())))
    l += 1
mapa = np.array(mapa)
print(f'Width: {N}, Height: {M}, Customers: {C}, Reply offices: {R}')
print(xc.shape, yc.shape, rc.shape)
print(xc,yc,rc)
print(mapa.shape)
print(mapa)
```

```python
plt.figure(figsize = (18,14))
sns.heatmap(data = mapa, annot = False)
```

```python
mapc = []
for i in range(M):
     mapc.append(mapa[i])
for i in range(C):
    x = xc[i]
    y = yc[i]
    mapc[y][x] = 3000
plt.figure(figsize = (16,16))
plt.axis('off')
plt.imshow(mapa)
```
